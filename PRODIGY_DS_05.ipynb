```python
accidents_severity = df1.groupby('Severity').count()['ID']
accidents_severity
```

```python
# Define colors for alternate slices
colors = ['yellow', 'green', 'orange', 'red']

fig, ax = plt.subplots(figsize=(8, 6), subplot_kw=dict(aspect="equal"))
label = [1, 2, 3, 4]
plt.pie(accidents_severity, labels=label, colors=colors,
        autopct='%1.1f%%', pctdistance=0.85)
circle = plt.Circle((0, 0), 0.5, color='white')
p = plt.gcf()
p.gca().add_artist(circle)
ax.set_title("Accident by Severity", fontdict={'fontsize': 14})
plt.tight_layout()
plt.show()
```
